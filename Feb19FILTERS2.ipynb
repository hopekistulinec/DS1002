```python
import numpy as np
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

def redfilter(photo):
    RED = np.copy(photo)
    RED[:,:,1:] = 0
    return RED
def greenfilter(photo):
    GREEN = np.copy(photo)
    GREEN[:,:,[0,2]] = 0
    return GREEN
def bluefilter(photo):
    BLUE = np.copy(photo)
    BLUE[:,:,:2] = 0
    return BLUE

url = input("Enter the URL of the .jpg file: ")

response = requests.get(url)
img = Image.open(BytesIO(response.content))

photo = np.array(img)

RED = redfilter(photo)
GREEN = greenfilter(photo)
BLUE = bluefilter(photo)

rgonly = np.copy(photo)
rgonly[:,:,2] = 0

rbonly = np.copy(photo)
rbonly[:,:,1] = 0

gbonly = np.copy(photo)
gbonly[:,:,0] = 0

plt.figure(figsize=(15, 10))

plt.subplot(2, 4, 1)
plt.imshow(photo)
plt.title('Original Image')

plt.subplot(2, 4, 2)
plt.imshow(RED)
plt.title('Red Filtered')

plt.subplot(2, 4, 3)
plt.imshow(GREEN)
plt.title('Green Filter')

plt.subplot(2, 4, 4)
plt.imshow(BLUE)
plt.title('Blue Filter')

plt.subplot(2, 4, 5)
plt.imshow(rgonly)
plt.title('Red and Green Only')

plt.subplot(2, 4, 6)
plt.imshow(rbonly)
plt.title('Red and Blue Only')

plt.subplot(2, 4, 7)
plt.imshow(gbonly)
plt.title('Green and Blue Only')

plt.show()
reconstructed_image = RED + GREEN + BLUE
if np.array_equal(photo, reconstructed_image):
    print("The original image is identical to reconstructed image.")
else:
    print("The original image is not identical to reconstructed image.")
```
